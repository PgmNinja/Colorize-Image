```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import io
import numpy as np
import os
import random
import tensorflow as tf
```

```python
original_width = 1920
original_height = 1080
target_width = 400
target_height = int(original_height * (target_width / original_width))
batch_size = 32
```

```python

file_path = './colorize-image-dataset/'
```

```python
X = []
all_files = os.listdir(file_path)
batch_size = 32


for i in range(0, len(all_files), batch_size):
    X_in_batch = []
    files_in_curr_batch = all_files[i:i+batch_size]
    
    for filename in files_in_curr_batch:
        loaded_image = load_img(file_path + filename).resize((target_width, target_height))
        X_in_batch.append(img_to_array(loaded_image))
        print(f'{filename} is loaded')
    X.extend(X_in_batch)
    
```

```python
import numpy as np

def update_in_batches(X, chunk_size):
    split = int(0.95 * len(X))
    num_chunks = int(np.ceil(split / chunk_size))

    Xtrain = np.empty_like(X[:split], dtype=float)

    for i in range(num_chunks + 1):
        start_idx = i * chunk_size
        end_idx = min((i + 1) * chunk_size, split)

        if start_idx >= end_idx:
            continue

        # Load a chunk of data
        chunk = np.array(X[start_idx:end_idx], dtype=float)

        # Normalize the chunk of data
        chunk = 1.0 / 255 * chunk

        # Assign the normalized chunk to the appropriate indices in Xtrain
        Xtrain[start_idx:end_idx] = chunk

    # Xtrain now contains the normalized data for the first 95% of the split
    return Xtrain
```

```python
Xtrain = update_in_batches(X, 50)
```

```python
from tensorflow.keras.layers import Conv2D, UpSampling2D, InputLayer, Conv2DTranspose
from tensorflow.keras.models import Sequential
```

```python
#Design the neural network
tf.keras.backend.clear_session()

model = Sequential()
model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(225, 400, 1)))
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same', strides=2))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(256, (3, 3), activation='relu', padding='same', strides=2))
model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(UpSampling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(UpSampling2D((2, 2)))
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(2, (3, 3), activation='tanh', padding='same'))
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

model.compile(optimizer='adam', loss='mse')
# model.save('colorize-image-model.h5')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
import numpy as np
```

```python
# Image transformer
datagen = ImageDataGenerator(
        shear_range=0.2,
        zoom_range=0.2,
        rotation_range=20,
        horizontal_flip=True)
```

```python
from skimage.color import rgb2lab, lab2rgb, rgb2gray, xyz2lab
import numpy as np

# Generate training data
batch_size = 50
def image_a_b_gen(batch_size):
    for batch in datagen.flow(Xtrain, batch_size=batch_size):
        lab_batch = rgb2lab(batch)
        X_batch = lab_batch[:,:,:,0]
        Y_batch = lab_batch[:,:,:,1:] / 128
        yield (X_batch.reshape(X_batch.shape+(1,)), Y_batch)
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir='/output')
model.fit(image_a_b_gen(batch_size), steps_per_epoch=1000, epochs=1, batch_size=10)
```
